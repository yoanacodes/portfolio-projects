```python
# filter just the Boroughs
deaths_by_borough = deaths_df[deaths_df.geotype_name == 'Borough']
```

```python
deaths_by_borough
```

```python
# plot line chart for the PM2.5 deaths by boroughs
fig, ax = plt.subplots(figsize=(40,15))
deaths_by_borough.groupby(['start_date','geo_place_name']).sum()['value'].unstack().plot(ax=ax, linewidth=3)

# place legend below the chart
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), fancybox=True, 
           shadow=True, ncol=5, borderaxespad=0., prop={'size': 25})
```

```python
# grouped bar chart showing the pm2.5 deaths by Borough

# plt.figure(figsize=(35,14))
# fig, ax = plt.subplots(figsize=(80,35))

deaths_by_borough.groupby(['start_date','geo_place_name']).sum()[['value']].unstack('geo_place_name').plot.bar()

plt.legend(loc='upper right')
```

```python
# Add a new column that associates each Congretional District and neighborhood with each borough

# 1 = Bronx ; 2 = Brooklyn ; 3 = Manhattan ; 4 = Queens ; 5 = Staten Island 

# copy column geo_join_id
data["borough_code"] = data["geo_join_id"]
data["borough_code"] = data["borough_code"].astype('str')

def changeCodeToName(x):
    if len(x) == 3:
        if x[0] == '1':
            x = 'Bronx'
        elif x[0] == '2':
            x = 'Brooklyn' 
        elif x[0] == '3':
            x = 'Manhattan' 
        elif x[0] == '4':
            x = 'Queens' 
        elif x[0] == '5':
            x = 'Staten Island'             
    else:
        x = False
    return x
    


data['borough_code'] = data['borough_code'].apply(lambda x: changeCodeToName(x))


data['borough_code'] = data['borough_code'].astype('category')

```
